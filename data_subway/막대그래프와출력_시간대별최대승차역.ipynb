```python
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.CSV','r',encoding='cp949')
data=csv.reader(f)
next(data)
next(data)

mx=[0]*24
mx_station=['']*24
for row in data:
    row[4:]=map(int, row[4:])
    for j in range(24):        
        a=row[2*j+4]
        if a>mx[j]:
            mx[j]=a
            if j<21:
                mx_station[j]=row[3]+'('+str(j+4)+')'
            else:
                mx_station[j]=row[3]+'('+str(j-20)+')'

#시간대 별 최대 승차역과 승객수 확인  
print(mx_station)
print(mx)

plt.rc('font', family = 'Malgun Gothic')
plt.style.use('ggplot')
plt.bar(range(24),mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```
