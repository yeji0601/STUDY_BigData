```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv','r',encoding='cp949')
data = csv.reader(f)
m=[]
f=[]
name = input('찾는 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+106].strip().replace(',','')))
            f.append(int(row[i+209].strip().replace(',','')))
        break;
        
plt.figure(figsize=(10, 5))
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+' 지역의 연령별 인구데이터')
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```
